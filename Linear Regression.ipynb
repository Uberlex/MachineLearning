Line of best fit: Minimized distance between both the line and the most dots on the graph

Determining Distance 
- Between points formula: $ \sqrt(x_2 - x_1)^2 + (y_2 - y_1)^2$
- General distance formula: d = $ \sqrt \sum (x_i - y_i)^2 $
- Minimizing the distance between points: **Mean Square Error (MSE)**
    - **Formula**: $$J(m,b) = \frac{1}{N} \sum^{N}_{i=1} (mx_i + b - y_i)^2  $$ 
    
    - $y_i - mx_i + b$ -> line, point subtract: distance between points -> minimization of just one point
    - squared number = elimination of negatives
    - 'sigma'= sum of all points
    - i = index to move through the points: mathematical 'for loop'
    - N = number of points 
    - m and b = slope and initial y- value to create the line -> parameters; weight values
    - y = y-value of a dot along the way that crosses the 
    - 2 = partial derivative
    - Used for regression tasks when referring to Loss Functions

### ***Stochastic Gradient Descent:***
---

Definition: A vector that lives where the weight values are drawn between the line and the point; defining parameters that revolve around the data a neural net will process to learn more about a specific task
- Gradient symbol: $ \nabla J$
- Analyzing each given data point one-by-one to determine a viable minimum with the most relative proportions to the overall data set's trajectory in a graphical interface

Application:
- Negative gradient = opposite direction of wherever the line is facing: greatest decrease
    - Navigation of direction upon the error surface
    
- Positive gradient = greatest increase

- Minimization of Loss Function

- Interpretation of Learning Rate

- Formula:$$\nabla J(m,b) = \left \langle \frac{\partial J}{\partial m},\frac{\partial J}{\partial b} \right \rangle $$
